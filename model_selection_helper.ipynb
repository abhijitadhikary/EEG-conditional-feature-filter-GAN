```python
import torch
```

```python
def print_results(filename):
    r = torch.load(filename)
    for i in range(len(r['acc_real'])):
        print('epoch %d'%r['epoch'][i])
        #print(r['acc_real'][i])
        print('disguised EEG: ')
        print(r['acc_fake'][i])
        print('============================================================================')
```

```python
print_results('cycada_stimulus_v2_ResNet34.pth')  #45 epoch
```

```python
print_results('cycada_combined_v2_ResNet34.pth') 
```
